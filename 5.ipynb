```
#Vectorized ridge regrression 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import cm

def upload_data(file_name):
  data = pd.read_excel(file_name, header=None)
  x1 = data.iloc[:,1]
  x2 = data.iloc[:,0]
  y = data.iloc[:,-1]
  mean1 = x1.mean()
  mean2 = x2.mean()
  mean3 = y.mean()
  std1 = x1.std()
  std2 = x2.std()
  std3 = y.std()
  x1 = (x1-mean1)/std1
  x2 = (x2-mean2)/std2
  y = (y-mean3)/std3

  X = np.column_stack((x1,x2))
  X = np.column_stack((np.ones(len(x1)),X))
  Y = np.array(y)
  return X,Y, mean1, mean2, mean3, std1, std2, std3

def upload_input(file_name, mean1, mean2, std1, std2):
  data = pd.read_excel(file_name, header=None)
  x1 = data.iloc[:,1]
  x2 = data.iloc[:,0]
  x1 = (x1-mean1)/std1
  x2 = (x2-mean2)/std2
  X1 = np.column_stack((x1,x2))
  X1 = np.column_stack((np.ones(len(x1)),X1))
  return X1

def upload_output(file_name):
  data = pd.read_excel(file_name, header=None)
  Y1 = np.array(data)
  return Y1

class RidgeRegression(object):

  def __init__(lambda1):
    self.lambda1 = lambda1

  def normal_bgd(self, X, Y, lambda1):
    X_T = np.matrix.transpose(X)
    self.theta = np.matmul(X_T,X) + (lambda1*np.identity(X.shape[1]))
    self.theta = np.linalg.pinv(self.theta)
    self.theta = np.matmul(self.theta, X_T) 
    self.theta = np.matmul(self.theta, Y)
    return self

  def predict(self, X1, Y1, mean3, std3, theta1):
    m = X1.shape[0]
    n = X1.shape[1]
    Y_predicted = np.zeros(X1.shape[0])
    mse1 = 0.0
    for i in range(m):
        x = X1[i]
        h = 0.0
        for j in range(n):
          h += theta1[j] * x[j]
        Y_predicted[i] = h

    Y_predicted = (Y_predicted*std3) + mean3 
    Y_predicted = np.array(Y_predicted)
    for i in range(m):
        error = Y_predicted[i] - Y1[i]
        mse1 += (error**2)

    mse1 = mse1 * (1.0/m)
    return mse1, Y_predicted

  def main():
  ##upload the .xlsx file with features and outputs and enter the name below
    training_file_name = 'training_feature_matrix.xlsx'
    X, Y, mean1, mean2, mean3, std1, std2, std3 = upload_data(training_file_name)
    test_input_file_name = 'test_feature_matrix.xlsx'
    X1 = upload_input(test_input_file_name, mean1, mean2, std1, std2)
    test_output_file_name = 'test_output_vector.xlsx'
    Y1 = upload_output(test_output_file_name)

    model = RidgeRegression(lambda1=0.25).normal_bgd(X,Y,lambda1)
    print("weights:", model.theta)

    mse1, Y_predicted = LinearRegression(lambda1=0.25).predict(X1,Y1,mean3, std3, model.theta)
    print("mse between actual and predicted output:", mse1)

main()
```

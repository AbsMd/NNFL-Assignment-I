```
#One Vs All Mutliclass Logistic Regression

import pandas as pd
import math 
import numpy as np

def get_column(data,c):
    column = []
    for i in range(len(data)):
      column.append(data[i][c])
    return column

def set(y):
    for i in range(len(y)):
        if(y[i]>=0.5):
            y[i] = 1
        if(y[i]<0.5):
            y[i] = 0
    return y

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def gradientDescent(w1,w2,w3,w4,w5,w6,w7,c,x,y):
    g1 = 0
    g2 = 0
    g3 = 0
    g4 = 0
    g5 = 0
    g6 = 0
    g7 = 0
    gc = 0
    yp = [0 for i in range(len(x))]
    for i in range(len(x)):
        yp[i] = (w1*x[i,0]) + (w2*x[i,1]) + (w3*x[i,2]) + (w4*x[i,3]) + (w5*x[i,4]) + (w6*x[i,5]) + (w7*x[i,6]) + c
        yp[i] = sigmoid(yp[i])
        g1 -= (y[i]-yp[i])*x[i,0]
        g2 -= (y[i]-yp[i])*x[i,1]
        g3 -= (y[i]-yp[i])*x[i,2]
        g4 -= (y[i]-yp[i])*x[i,3]
        g5 -= (y[i]-yp[i])*x[i,4]
        g6 -= (y[i]-yp[i])*x[i,5]
        g7 -= (y[i]-yp[i])*x[i,6]
        gc -= (y[i]-yp[i])
    w1 -= 0.1*g1
    w2 -= 0.1*g2
    w3 -= 0.1*g3
    w4 -= 0.1*g4
    w5 -= 0.1*g5
    w6 -= 0.1*g6
    w7 -=  0.1*g7
    c = c - 0.1*gc
    return w1,w2,w3,w4,w5,w6,w7,c

def linearRegression(x,y):
    w1 = 0
    w2 = 0
    w3 = 0
    w4 = 0
    w5 = 0
    w6 = 0
    w7 = 0
    c = 0
    iterations = 2000
    i = 0
    while(i<iterations):
        w1,w2,w3,w4,w5,w6,w7,c = gradientDescent(w1,w2,w3,w4,w5,w6,w7,c,x,y)
        i = i+1
    return w1,w2,w3,w4,w5,w6,w7,c

#Normalization
def normalize(data):
    normalized_data = data
    for i in range(7):
        maxval = np.amax(get_column(data,i))
        minval = np.amin(get_column(data,i))
        for j in range(len(data)):
            normalized_data[j][i] = (data[j][i]-minval)/((maxval-minval)+0.05)
    return normalized_data


#Data
data = pd.read_excel('data4.xlsx',header=None)
data = data.sample(frac=1)
data = np.asarray(data)
y = data[:,-1]
data = normalize(data)
x = data[:,:-1]


#Split into testing and training sets
train_size = int(0.6 * len(x))
x_training = x[:train_size]
x_test = x[train_size:]
y_training = y[:train_size]
y_test = y[train_size:]

y1_training = [1 for i in range(len(y_training))]
y2_training = [1 for i in range(len(y_training))]
y3_training = [1 for i in range(len(y_training))]
for i in range(len(y_training)):
    if(y_training[i] != 1):
        y1_training[i] = 0
    if(y_training[i] != 2):
        y2_training[i] = 0
    if(y_training[i] != 3):
        y3_training[i] = 0

x = x_test

#n-predictions
w1,w2,w3,w4,w5,w6,w7,c = linearRegression(x_training,y1_training)
y_p1 = [0 for i in range(len(x))]
for i in range(len(x)):
    y_p1[i] = (w1*x[i,0]) + (w2*x[i,1]) + (w3*x[i,2]) + (w4*x[i,3]) + (w5*x[i,4]) + (w6*x[i,5]) + (w7*x[i,6]) + c
    y_p1[i] = sigmoid(y_p1[i])
y_p1 = set(y_p1)

w1,w2,w3,w4,w5,w6,w7,c = linearRegression(x_training,y2_training)
y_p2 = [0 for i in range(len(x))]
for i in range(len(x)):
    y_p2[i] = (w1*x[i,0]) + (w2*x[i,1]) + (w3*x[i,2]) + (w4*x[i,3]) + (w5*x[i,4]) + (w6*x[i,5]) + (w7*x[i,6]) + c
    y_p2[i] = sigmoid(y_p2[i])
y_p2 = set(y_p2)

w1,w2,w3,w4,w5,w6,w7,c = linearRegression(x_training,y3_training)
y_p3 = [0 for i in range(len(x))]
for i in range(len(x)):
    y_p3[i] = (w1*x[i,0]) + (w2*x[i,1]) + (w3*x[i,2]) + (w4*x[i,3]) + (w5*x[i,4]) + (w6*x[i,5]) + (w7*x[i,6]) + c
    y_p3[i] = sigmoid(y_p3[i])
y_p3 = set(y_p3)

c_value = [0 for i in range(len(y_test))]
for i in range(len(y_test)):
    if (y_p1[i] == 1):
        c_value[i] = 1.0
    if (y_p2[i] == 1):
        c_value[i] = 2.0
    if (y_p3[i] == 1):
        c_value[i] = 3.0


for i in range(len(c_value)):
    if (c_value[i] == 0):
        c_value[i] = 'None'
y_actual = pd.Series(y_test, name='Actual')
y_pred = pd.Series(c_value, name='Predicted')
confmat = pd.crosstab(y_actual, y_pred)
print(confmat)

confmat = np.asarray(confmat)
accuracy = (confmat[0][0] + confmat[1][1] + confmat[2][2])/sum(sum(confmat))
accuracy_1 = confmat[0][0]/sum(confmat[0])
accuracy_2 = confmat[1][1]/sum(confmat[1])
accuracy_3 = confmat[2][2]/sum(confmat[2])
print('Overall Accuracy : ' + str(accuracy))
print('Accuracy of class 1 : ' + str(accuracy_1))
print('Accuracy of class 2 : ' + str(accuracy_2))
print('Accuracy of class 3 : ' + str(accuracy_3))
```
